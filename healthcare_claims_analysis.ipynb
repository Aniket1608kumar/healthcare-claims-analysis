# Healthcare Claims Data Analysis using Python
This project analyzes healthcare claim data using Python. The goal is to explore claim patterns, department distributions, approval rates, and claim amount trends.

## Import Libraries
This section imports the required Python libraries such as Pandas, NumPy, and Matplotlib for data manipulation and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Dataset
This section generates a synthetic healthcare claims dataset containing department names, claim amounts, and approval status.

```python
np.random.seed(42)

data = {
    "Department": np.random.choice(["Medical","Dental","Vision"],100),
    "Claim_Amount": np.random.randint(200,2000,100),
    "Approved": np.random.choice([0,1],100)
}

df = pd.DataFrame(data)
df
```

## Data Overview
This section displays the first few rows and summary statistics of the dataset to understand its structure and key characteristics.

```python
#overview 
print(df)
print()


print("\n top 5 value of DF \n ",df.head())
print()

print("\n bottom 5 value of DF \n ",df.tail())
print()

print("\n describe  DF \n ",df.describe().astype(int))
print()
```

## Department Distribution
This section analyzes the number of claims submitted in each department and visualizes the distribution using a bar chart.

```python
 #  Visualization  -- (Department claim count:)


ax = df["Department"].value_counts().plot(kind="bar")
ax.bar_label(ax.containers[0])
plt.title("Claims per Department")
plt.show()
```

## Average Claim Amount
This section calculates the average claim amount for each department to compare claim costs across different healthcare services

```python
df.groupby("Department")["Claim_Amount"].mean().astype(int)
```

## Approval Rate
This section calculates the percentage of approved and rejected claims to understand the overall approval rate in the dataset.

```python
df["Approved"].value_counts(normalize=True)*100
```

## Claim Amount Distribution
This section visualizes the distribution of claim amounts using a histogram to identify common claim ranges and variability.

```python
# Claim amount distribution:

df["Claim_Amount"].plot(kind="hist", bins=30)
#plt.hist(df["Claim_Amount"])
plt.title(" Claim amount distribution")
plt.show()
```
